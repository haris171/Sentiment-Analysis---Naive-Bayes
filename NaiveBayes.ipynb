```
from collections import Counter
import pandas as pd
import numpy as np
import string
import array
import math
import os
import re 
```

#Extracting Data into Variables

```
train_pos_files = os.listdir("Dataset/train/pos") # Saving file names into a list
train_neg_files = os.listdir("Dataset/train/neg")
```

```
test_pos_files = os.listdir("Dataset/test/pos")
test_neg_files = os.listdir("Dataset/test/neg")
```

```
with open('Dataset/stop_words.txt', encoding = "ISO-8859-1") as s:
  stop = s.readlines()
```

```
stop_words1 = []   # Storing negative, stop and positive words into respective lists

for z in stop:
  word = z.strip()
  stop_words1.append(word)
```

```
def file_read(filename,directory):    #Reading file and doing pre-processing on the data and counting positive and negative words

  with open(os.path.join(directory, filename)) as n:
    data = n.readlines()
  
  
  data = data[0].lower()
  unprocessed_data = data     # This is like lower cased reviews, for sklearn part 2

  data = data.split()
  data_upd = list()

  for word in data:
    word = re.sub('[@&%*/-/+/.,!0-9]*', '', word)
    item = word.strip(string.punctuation) #Remove punctuations
    if item not in stop_words1 and item is not '': 
      data_upd.append(item)

  return data_upd, unprocessed_data
```

# Reading Files 

```
x0 = []   #Creating training 
t_train = []
pos_list = []
reviews = []

for filename in train_pos_files:
  p, r = file_read(filename,"Dataset/train/pos")
  x0.append(p)
  reviews.append(r)
  pos_list.append(p)
  t_train.append(1)


print(len(x0))
```

```
x1 = []
neg_list = []
for filename in train_neg_files:
  n, rn = file_read(filename,"Dataset/train/neg")
  x0.append(n)      
  reviews.append(rn)
  neg_list.append(n)
  t_train.append(0)

print(len(x0))
```

```
#Testing data 

tx1 = []
t_test = []
reviews_test = []
for filename in test_pos_files:
  p, rt = file_read(filename,"Dataset/test/pos")
  tx1.append(p)
  reviews_test.append(rt)
  t_test.append(1)

for filename in test_neg_files:
  n, rtn = file_read(filename,"Dataset/test/neg")
  tx1.append(n)
  reviews_test.append(rtn)
  t_test.append(0)

testX = tx1
```

```
def make_dictionary(x):   #Making My list of words
  voc = []
  for word in x:
    if word not in voc: 
      voc.append(word)
  return voc
```

```
def make_vocab(X):
  vocab = []
  v = []
  voc = []

  for item in X: #Making my vocabalary using the above function, i concatenate list of words
    v = make_dictionary(item)
    vocab.append(v)

  for i in range(len(vocab)):
    for word in vocab[i]:
      voc.append(word)

  return voc
```

```
vocab_train = make_vocab(trainX)
# print(len(vocab_train))
# vocab = dict()
vocab = Counter(vocab_train)
len(vocab.keys())
```

```
v_pos = make_vocab(pos_list) # Making positive and negative vocabs so its easy to calculate the liklihood, as it is class based
vocab_pos = Counter(v_pos)
len(vocab_pos.keys())
```

```
v_neg = make_vocab(neg_list)
vocab_neg = Counter(v_neg)
len(vocab_neg.keys())
```

# Functions for Naive Bayes

```
def train_Naive_Bayes(D,C,V,V_pos,V_neg):
  
  log_L = dict()
  log_p = dict()
  Nc = 12500
  Ndoc = len(trainX)
  temp_pos = dict()
  temp_neg = dict()
  count_pos = sum(V_pos.values())     # Total counts of Pos
  count_neg = sum(V_neg.values())     # Total counts of neg
  card_V = len(V.keys())              # Cardinality of V
  
  for c in C:
    log_p[c] = np.log(Nc/Ndoc)  # Calculating the prior
    for word in V.keys():
      if c == 'positive':
        temp_pos[word] = np.log((V_pos[word] + 1)/(count_pos + card_V)) # formula for each seperate class
        log_L[c] = temp_pos
      elif c == 'negative':
        temp_neg[word] = np.log((V_neg[word] + 1)/(count_neg + card_V))
        log_L[c] = temp_neg

  return log_p, log_L
```

```
def test_Naive_Bayes(test_doc,log_p,log_L,C,V):

  sumt = dict()
  for c in C:
    sumt[c] = log_p[c]
    for i in range(len(test_doc)):
      word = test_doc[i]
      if word in V:
        if c == 'positive':
          sumt[c] = sumt[c] + log_L[c][word]    
          a = sumt[c]
        elif c== 'negative':
          sumt[c] = sumt[c] + log_L[c][word]
          b = sumt[c]
    
  if a > b:     # Returning the class based on which returns the highhest probabibilty
    return 1
  elif a < b:
    return 0
```

```
def confusion_matrix_1(test_pred,t_test):
  tp = 0
  tn = 0
  fp = 0
  fn = 0
  for i in range(len(t_test)):
    if test_pred[i] == t_test[i]:
      if test_pred[i] == 1: 
        tp = tp + 1
      elif test_pred[i] == 0:
        tn = tn + 1 
    elif test_pred[i] != t_test[i]:
      if test_pred[i] == 1:
        fp = fp + 1
      elif test_pred[i] == 0:
        fn = fn + 1
  return tp,tn,fp,fn
```

# Testing Naive Bayes

```
classes = ['positive','negative']
P , L = train_Naive_Bayes(trainX,classes,vocab,vocab_pos,vocab_neg)
```

```
test_pred = []
for i in range(len(testX)):       # Testing each tweet 1 by 1
  pred_x = test_Naive_Bayes(testX[i],P,L,classes,vocab.keys())
  test_pred.append(pred_x)
```

```
tp, tn, fp, fn = confusion_matrix_1(test_pred,t_test)
acc = (tp+tn)/(tp+tn+fp+fn)
print('Accuracy of the Naive Bayes is = ', acc*100, '%')
```

#Part2 Using Sklearn

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```
vector = CountVectorizer(stop_words=stop_words1) # Removing stop words, list is provided

X = vector.fit_transform(reviews)
T = vector.transform(reviews_test)

clf = MultinomialNB()
clf.fit(X,t_train)
pred = clf.predict(T)
```

```
accu = accuracy_score(t_test,pred)
print('Accuracy of the Naive Bayes using sklearn is = ', accu*100, '%')
```

```
C = confusion_matrix(t_test,pred)
print('Confusion Matrix of the Naive Bayes using sklearn is')
print(C)
```
